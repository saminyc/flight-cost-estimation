<h2>Preprocessing</h2>

```python
df = df.drop(['Unnamed: 0', 'flight'], axis=1)
```

```python
df
```

```python
# Class to binary and stops to numeric
```

```python
df['class'] = df['class'].apply(lambda x: 1 if x == 'Business' else 0)
```

```python
df['stops'] = pd.factorize(df['stops'])[0]
```

```python
columns_to_encode = ['source_city', 'destination_city', 'departure_time', 'arrival_time']
prefixes = ['source', 'dest', 'departure', 'arrival']

for col, prefix in zip(columns_to_encode, prefixes):
  encoded_col = pd.get_dummies(df[col], prefix=prefix)
  df = pd.concat([df, encoded_col], axis=1)
  df = df.drop(col, axis=1)
```

```python
df
```

```python
df = df.join(pd.get_dummies(df.airline, prefix='airline')).drop('airline',axis=1)
```

```python
df = df.replace({True: 1, False: 0})
```

```python
df
```
